```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carrega os dados
df = pd.read_csv("../data/processed/temperatura_rj_tratada.csv")


# Gráfico de linha da média mensal
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="month", y="temperature_2m_avg", marker="o")
plt.title("Temperatura Média Mensal no RJ - 2023")
plt.xticks(rotation=45)
plt.ylabel("Temperatura Média (°C)")
plt.grid(True)
plt.tight_layout()
plt.show()

plt.savefig("grafico_temperatura.png")

```
